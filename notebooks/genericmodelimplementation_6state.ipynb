```python
#### Create model instance
#Full Lists
blank_list = [blank1, blank2, blank3, blank4]
edtaca_list = [caedta1, edtaca1, edtaca2, edtaca3, edtaca4, edtaca5, edtaca6]
prot_list = [prot1, prot2, prot3, prot4, prot5, prot6, reprot1, reprot2]

#Combine experiment types into one list
expt_list = blank_list + edtaca_list + prot_list


# Read the model specification from file
spec_file_path = r"C:\Users\willi\linkage\src\linkage\symbolic\model_specs\SixStateEDTA.txt"

# Read spec
with open(spec_file_path, 'r') as f:
    model_spec = f.read()

# Create GlobalModel with spec
gm = linkage.GlobalModel(
    model_name="GenericBindingModel",
    model_spec=model_spec,
    expt_list=expt_list
)

#Setup dataprob
f = dataprob.setup(gm.model_normalized,
                   method="ml",
                   vector_first_arg=True,
                   fit_parameters=gm.parameter_names)
```

```python
f.param_df
```

```python
# Create a dictionary to hold the complete configuration for each parameter.
# This makes it easy to see all settings for a given parameter in one place.
param_configs = {

    # --- Equilibrium Constants (lnK) ---
    # Since A is favored over I, KI = [I]/[A] << 1, so ln(KI) must be negative.
    "KI": {"guess": -12.0, "lower_bound": -12,  "upper_bound": -5,   "fixed": True},
    
    # High-affinity Ca++ binding sites
    "K1": {"guess": 6, "lower_bound": 5,    "upper_bound": 15,   "fixed": False},
    "K2": {"guess": 6, "lower_bound": 5,    "upper_bound": 15,   "fixed": False},

    # Low-affinity Ca++ binding sites
    "K3": {"guess": 4,  "lower_bound": 1,    "upper_bound": 8,   "fixed": False},
    "K4": {"guess": 4,  "lower_bound": 1,    "upper_bound": 8,   "fixed": False},
    
    # EDTA binding constant
    "KE": {"guess": 14.7,"lower_bound": 12.0,"upper_bound": 16.4,"fixed": False},

    # --- Enthalpies (in kCal/mol) ---
    # NOTE: GlobalModel now scales these by 1000 internally. Input small numbers!
    # Assumed isoenthalpic for the inactive->active transition
    "dH_I": {"guess": 0, "fixed": True}, 
    
    # Binding dH should be within a physical range (-7 kCal, not -7000)
    "dH_1": {"guess": -7.0, "lower_bound": -12.0, "upper_bound": -4.0},
    "dH_2": {"guess": -7.0, "lower_bound": -12.0, "upper_bound": -4.0},
    "dH_3": {"guess": -0.1, "lower_bound": -3.0, "upper_bound": 5.0},
    "dH_4": {"guess": -0.1, "lower_bound": -3.0, "upper_bound": 5.0},

    # EDTA binding enthalpy (-10.985 kCal)
    "dH_E": {"guess": -10.985, "lower_bound": -11.035, "upper_bound": -10.925, "fixed": False},

    # --- Nuisance Parameters: Dilution (in kCal/mol) ---
    # -482 cal -> -0.482 kCal
    "nuisance_dil_CT": {"guess": -0.482, "lower_bound": -1.0, "upper_bound": 1.0, "fixed": True},
    "nuisance_dil_ET": {"guess": -0.056, "lower_bound": -1.0, "upper_bound": 1.0, "fixed": True},
}

# Apply the configurations to the parameter DataFrame
for param_name, settings in param_configs.items():
    if param_name in f.param_df.index:
        # Use .get() to avoid errors if a key (like 'fixed') is not specified
        for key, value in settings.items():
            f.param_df.loc[param_name, key] = value
    else:
        print(f"Warning: Parameter '{param_name}' from config not in model.")

# --- Nuisance Parameters: Experimental Fudge Factors ---
fudge_params = [name for name in f.param_df.index if 'nuisance_expt' in name]
for param_name in fudge_params:
    f.param_df.loc[param_name, 'guess'] = 1.0
    f.param_df.loc[param_name, 'fixed'] = True  # Recommended: Float these!
    f.param_df.loc[param_name, 'lower_bound'] = 0.9
    f.param_df.loc[param_name, 'upper_bound'] = 1.1

# --- Set Prior Means from Guesses ---
# This uses the initial guess for each parameter as the mean for its prior
# distribution in the Bayesian model.
f.param_df['prior_mean'] = f.param_df['guess']

# --- Set Prior Standard Deviations from Means ---
# Set a fractional value for the standard deviation relative to the mean.
# 0.5 means the std dev will be 50% of the absolute value of the guess.
proportion = 0.5
f.param_df['prior_std'] = proportion * abs(f.param_df['prior_mean'])

# Handle the edge case where the guess is 0. A std dev cannot be 0.
# We set it to a reasonable default value, like 1.0, for these cases.
f.param_df.loc[f.param_df['prior_mean'] == 0, 'prior_std'] = 1.0

f.param_df
```

```python
### ML FITTER FUNCTION CALL (Requires method="ml" in the dataprob fitter setup)

f.fit(
    y_obs=gm.y_obs_normalized,
    y_std=gm.y_std_normalized,

    # --- Core Arguments for the Optimizer ---
    method='trf',          # Trust Region Reflective is good for bounded problems.

    # --- Tolerances ---
    # Loosen ftol/gtol slightly to handle flat regions, keep xtol tight.
    ftol=1e-9,             # Termination by change in cost function.
    xtol=1e-9,             # Termination by change in parameters.
    gtol=1e-9,             # Termination by norm of the gradient.

    # --- Scaling and Robustness ---
    x_scale='jac',         # Crucial for problems where parameters have very different
                           # magnitudes. Let the Jacobian estimate the scales.
    loss='linear',         # Standard least-squares. Change to 'soft_l1' if you
                           # suspect outliers in your data.

    # --- Number of function evaluations ---
    max_nfev=20,

    # --- Verbosity ---
    verbose=2              # Keep this at 2 to see the step-by-step progress
                           # of the optimization.
)
```

```python
### Emcee FITTER FUNCTION CALL (Requires method="emcee" in the dataprob fitter setup)


f.fit(
    y_obs=gm.y_obs_normalized,
    y_std=gm.y_std_normalized,

    # Number of walkers to explore parameter space. Should be <2 times the number of fit parameters.
    num_walkers=25,

    # Initial number of steps for each walker before checking convergence.
    num_steps=50,

    # Use a preliminary ML fit to find a good starting position for the walkers.
    use_ml_guess=False,

    # The sampler will automatically try to extend the run this many times to meet convergence criteria.
    max_convergence_cycles=3,
    
    # Fraction of initial steps to discard from each walker for the final analysis.
    burn_in=0.2,
)


### 0.5s per walker per step per experiment??
```

```python
### PyMC FITTER FUNCTION CALL (Requires method="pymc" in the dataprob fitter setup)

f.fit(
    y_obs=gm.y_obs_normalized,
    y_std=gm.y_std_normalized,

    # Number of independent chains to run. Conceptually similar to walkers.
    chains=4,

    # Number of samples to generate and keep from each chain.
    draws=30,

    # Number of "burn-in" or "warmup" steps to tune the sampler. These are discarded.
    tune=10,
)
```

```python
#Multicolor graphing

style = {"s":50,
         "facecolor":"none",
         "edgecolor":"black"}
err_style = {"lw":0,
             "elinewidth":1,
             "capsize":2}

orange_list = ['#FFEEDD', '#FFD6AA', '#FFB366', '#FF9933', '#FF8000', '#FF6600', '#FF4400', '#CC3300', '#992200', '#FF0000'] 
purple_list = ['#F2E6FF', '#E0B3FF', '#CC80FF', '#B84DFF', '#A31AFF', '#8800E6', '#6600B3', '#440080', '#2B0052', '#1A0033']
green_list = ['#E8FFE8', '#C1FFC1', '#9AFF9A', '#74FF74', '#4DFF4D', '#26FF26', '#00E600', '#00B300', '#008000', '#004D00']

edtaca_length = len(edtaca_list)
prot_length = len(prot_list)
blank_length = len(blank_list)

color_order = green_list[0:blank_length] + orange_list[0:edtaca_length] + purple_list[0:prot_length]

fig, ax = plt.subplots(1,figsize=(6,6))

out_df = gm.as_df.copy()
y_calc = gm.model(np.array(f.fit_df["estimate"]))

for i in np.unique(out_df.expt_id):
    
    style["edgecolor"] = color_order[i]
    err_style["color"] = color_order[i]

    mask = out_df["expt_id"] == i
    this_df = out_df.loc[mask,:]

    
    x_values = np.cumsum(this_df["injection"])
    y_values = np.array(this_df["y_obs"])
    y_err = np.array(this_df["y_std"])/np.mean(this_df["injection"])
    this_y_calc = y_calc[mask]/this_df["injection"]

    y_values = y_values/this_df["injection"]
    
    ax.scatter(x_values,y_values,**style)
    ax.errorbar(x=x_values,
                y=y_values,
                yerr=y_err,
                **err_style)

    ax.plot(x_values,this_y_calc,'-',color=color_order[i])

ax.set_ylim((-100,10))

plt.xlabel("injection")
plt.ylabel("heat")
```
