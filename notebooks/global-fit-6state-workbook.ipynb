```python
#### Create model instance
#blank_list = [blank1, blank2]
#edtaca_list = [edtaca1, edtaca2, edtaca3, edtaca4, edtaca5, edtaca6, edtaca7, caedta1]
prot_list = [prot1, prot2, prot3, prot4, prot5, prot6]

#Combine experiment types into one list
#expt_list = blank_list + edtaca_list + prot_list

gm = linkage.GlobalModel(model_name="SixStateEDTA",
                         expt_list=prot_list)

f = dataprob.setup(gm.model_normalized,
                   vector_first_arg=True,
                   method="ml",
                   fit_parameters=gm.parameter_names)
```

```python
## Show empty parameter df
f.param_df
```

```python
## Assign parameter attributes

#Equilibrium Constant Guesses
f.param_df.loc["KI","guess"] = -4.6
f.param_df.loc["KI","upper_bound"] = -2
f.param_df.loc["KI","lower_bound"] = -10
f.param_df.loc["KI","fixed"] = True

f.param_df.loc["KE","guess"] = 16.4
f.param_df.loc["KE","upper_bound"] = 18.5
f.param_df.loc["KE","lower_bound"] = 15.0
f.param_df.loc["KE","fixed"] = True

f.param_df.loc["K1","guess"] = 7
f.param_df.loc["K1","upper_bound"] = 7.93
f.param_df.loc["K1","lower_bound"] = 6.14
f.param_df.loc["K1","fixed"] = False

f.param_df.loc["K2","guess"] = 12.7
f.param_df.loc["K2","upper_bound"] = 13.48
f.param_df.loc["K2","lower_bound"] = 11.95
f.param_df.loc["K2","fixed"] = False

f.param_df.loc["K3","guess"] = 7
f.param_df.loc["K3","upper_bound"] = 10
f.param_df.loc["K3","lower_bound"] = 2
f.param_df.loc["K3","fixed"] = False

f.param_df.loc["K4","guess"] = 7
f.param_df.loc["K4","upper_bound"] = 10
f.param_df.loc["K4","lower_bound"] = 2
f.param_df.loc["K4", "fixed"] = False


# ### Nuisance Param Guesses
f.param_df.loc["nuisance_dil_CT","guess"] = -149
f.param_df.loc["nuisance_dil_CT","upper_bound"] = -148
f.param_df.loc["nuisance_dil_CT","lower_bound"] = -150
f.param_df.loc["nuisance_dil_CT","fixed"] = True

f.param_df.loc["nuisance_dil_ET","guess"] = -6.1
f.param_df.loc["nuisance_dil_ET","upper_bound"] = -6.0
f.param_df.loc["nuisance_dil_ET","lower_bound"] = -6.2
f.param_df.loc["nuisance_dil_ET","fixed"] = True


# Get all parameter names containing 'nuisance_expt' and 'ET_fudge'
fudge_params = [col for col in f.param_df.index if 'nuisance_expt' in col]

# Link all fudge parameters (except 0) to the first one
for param in fudge_params:
        f.param_df.loc[param, 'guess'] = 1.05
        f.param_df.loc[param, 'fixed'] = True
        f.param_df.loc[param, 'lower_bound'] = -2
        f.param_df.loc[param, 'upper_bound'] = 2

# # ### Enthalpy Guesses

f.param_df.loc["dH_I","guess"] = 0
f.param_df.loc["dH_I","fixed"] = True

f.param_df.loc["dH_E","guess"] = -10852
f.param_df.loc["dH_E","upper_bound"] = -10800
f.param_df.loc["dH_E","lower_bound"] = -10900
f.param_df.loc["dH_E","fixed"] = True

f.param_df.loc["dH_1","guess"] = 100
f.param_df.loc["dH_1","upper_bound"] = 10000
f.param_df.loc["dH_1","lower_bound"] = 0

f.param_df.loc["dH_2","guess"] = 100
f.param_df.loc["dH_2","upper_bound"] = 10000
f.param_df.loc["dH_2","lower_bound"] = 0

f.param_df.loc["dH_3","guess"] = 100
f.param_df.loc["dH_3","upper_bound"] = 10000
f.param_df.loc["dH_3","lower_bound"] = 0

f.param_df.loc["dH_4","guess"] = 100
f.param_df.loc["dH_4","upper_bound"] = 10000
f.param_df.loc["dH_4","lower_bound"] = 0


# f.param_df.loc["","parent"] = ''
```

```python
## Check parameter assignments
f.param_df
```

```python
# Run fit

f.fit(
      y_obs=gm.y_obs_normalized,
      y_std=gm.y_std_normalized,
      #max_convergence_cycles=1,
      #use_ml_guess=False,
      #num_steps=800,
      #num_walkers=200, # number of markov chains to use in the analysis, default=100       
      method='trf',  # Algorithm to use for optimization
      jac='3-point',  # Method for computing the Jacobian matrix
      ftol=1e-15,  # Tolerance for termination by the change of the cost function
      xtol=1e-12,  # Tolerance for termination by the change of the independent variables
      gtol=1e-12,  # Tolerance for termination by the norm of the gradient
      x_scale='jac',  # Scaling of the variables
      loss='linear',  # Loss function for dealing with outliers
      f_scale=0.1,  # Soft margin between inlier and outlier residuals
      max_nfev=25,  # Maximum number of function evaluations
      verbose=2  # Level of algorithm's verbosity
     )
```

#### Plot results

```python
style = {"s":50,
         "facecolor":"none",
         "edgecolor":"black"}
err_style = {"lw":0,
             "elinewidth":1,
             "capsize":2}

orange_list = ['#FFEEDD', '#FFD6AA', '#FFB366', '#FF9933', '#FF8000', '#FF6600', '#FF4400', '#CC3300', '#992200', '#FF0000'] 
purple_list = ['#F2E6FF', '#E0B3FF', '#CC80FF', '#B84DFF', '#A31AFF', '#8800E6', '#6600B3', '#440080', '#2B0052', '#1A0033']
green_list = ['#E8FFE8', '#C1FFC1', '#9AFF9A', '#74FF74', '#4DFF4D', '#26FF26', '#00E600', '#00B300', '#008000', '#004D00']

# edtaca_length = len(edtaca_list)
prot_length = len(prot_list)
# blank_length = len(blank_list)

#color_order = green_list[0:blank_length] + orange_list[0:edtaca_length] + purple_list[0:prot_length]
color_order = purple_list[0:prot_length]

fig, ax = plt.subplots(1,figsize=(6,6))

out_df = gm.as_df.copy()
y_calc = gm.model(np.array(f.fit_df["estimate"]))

for i in np.unique(out_df.expt_id):
    
    style["edgecolor"] = color_order[i]
    err_style["color"] = color_order[i]

    mask = out_df["expt_id"] == i
    this_df = out_df.loc[mask,:]

    
    x_values = np.cumsum(this_df["injection"])
    y_values = np.array(this_df["y_obs"])
    y_err = np.array(this_df["y_std"])/np.mean(this_df["injection"])
    this_y_calc = y_calc[mask]/this_df["injection"]

    y_values = y_values/this_df["injection"]
    
    ax.scatter(x_values,y_values,**style)
    ax.errorbar(x=x_values,
                y=y_values,
                yerr=y_err,
                **err_style)

    ax.plot(x_values,this_y_calc,'-',color=color_order[i])
    

plt.xlabel("injection")
plt.ylabel("heat")

f.fit_df
```
